```python
import json
import re
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
```

```python
data = []
with open('location.json') as json_file:
    for line in json_file:
        data.append(json.loads(line))
```

### Return all addresses that start with a number (returning just the address)

```python
num_addresses = df['payload.address'][df['payload.address'].str.contains('^ *\d.*$', na=False)]
```

```python
num_addresses.sample(5)
```

### Return all rows with addresses that don't contain a number (return the entire row)

```python
non_num_records = df[df['payload.address'].str.contains('^[^0-9]+$', na=False)]
```
